```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

We use the Radial Velocity measurements carried out by Gajdo≈° et al. using 2 meter Perek telescope.

```python
rad_vel_data = pd.read_csv("E:/Projects/eclipsing_binaries_project/data/TIC_158794976_radial_velocity_data.csv")
```

```python
x = rad_vel_data["Phase"]
y = rad_vel_data["RV (km/s)"]
y_err = rad_vel_data["Error (km/s)"]

y_max = np.max(y)
y_min = np.min(y)
print(y_max,y_min)
```

```python
plt.errorbar(x=x, y=y, yerr=y_err,fmt=".k")
plt.xlabel("Phase")
plt.ylabel("Radial velocity (km/s)")
plt.show()
```

So Semi amplitude (K1) is approximately equal to:

=    (y_max - y_min)/2 

=    (39.69 -(-8.34))/2

=    24.015  km/ s



This is pretty close to the value (22.28 ¬± 2.71 km/s) predicted by Gajdo≈° et al, considering the approximation used in unmodeled RV curve
