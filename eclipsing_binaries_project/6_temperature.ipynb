Instead of building spectroscopic model of the system's photosphere to determine its temperature, we will use the blackbody approximation of the stars.

At this point, we will consider the values of R1 and R2, i.e, radii of the two binaries as determined by Gajdo≈° et al.

```python
import lightkurve as lk
import astropy.units as u
import matplotlib.pyplot as plt
```

```python
R1 = 2.2 * u.R_sun
R2 = 0.9 * u.R_sun
```

We would also need depth of the eclipse for this calculation.

```python
lc = lk.search_lightcurve("TIC 158794976")[0].download()     
lc.scatter() 
plt.xlim(1689,1693) 

plt.axhline(y=17500,color = 'blue')
plt.text(x=1690.5,y= 17550, s="Reference line",color='red',ha= "left")

plt.axhline(y=16280,color='red')
plt.text(x=1689,y= 16300, s="Secondary eclipse depth",color='red',ha= "right")

plt.axhline(y=16100,color='green')
plt.text(x=1689,y= 16120, s="Primary eclipse depth",color='red',ha= "right")

plt.show()
```

```python
primary_depth = 17500 - 16100
secondary_depth = 17500 - 16280
```

We know that ratio of primary_depth to secondary_depth is equal to the ratio (T1/T2)^4, considering the blackbody approximation by fourth power emission and temperature relation given by Stefan's Law.

```python
T1_T2_ratio = (primary_depth/secondary_depth)**(1/4)
T1_T2_ratio
```

As per Gajdo≈° et al, the ratio of temperatures, i.e, T1 / T2 comes out to be:
