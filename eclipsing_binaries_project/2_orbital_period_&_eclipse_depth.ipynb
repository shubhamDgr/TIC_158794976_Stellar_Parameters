```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
lc = lk.search_lightcurve("TIC 158794976")[0].download()     
lc.scatter() 
plt.xlim(1689,1693)                                      
plt.axvline(x=1689.82)                        # To measure period
plt.axvline(x=1690.6)
plt.axvline(x=1691.36)
plt.show()
```

Period 1 = 1690.6-1689.82= 0.78

Period 2 = 1691.36-1690.6= 0.76

Mean orbital period= 0.77 days = 6.6528 x 10^4 sec 

```python
bls = lc.to_periodogram("bls",frequency_factor=10, duration=0.05)       

lc.fold(bls.period_at_max_power.value, bls.transit_time_at_max_power).scatter()     

plt.axvline(x=-0.06,ymin=0,ymax=17000, color='red')         # plots vertical line 
plt.text(x=-0.075,y=17500,s="t1",color='red',ha='left')     # writes text adjacent to line

plt.axvline(x=-0.012,ymin=0,ymax=17000, color='blue')
plt.text(x=-0.022,y=17500,s="t2",color='blue',ha='left')

plt.axvline(x=0.007,ymin=0,ymax=17000, color='blue')
plt.text(x=0.012,y=17500,s="t3",color='blue',ha='left')

plt.axvline(x=0.057,ymin=0,ymax=17000, color='red')
plt.text(x=0.06,y=17500,s="t4",color='red',ha='left')
plt.show()
```

t4- t1 = 0.117 * period = 7783 sec

This is the duration for which the two stars overlap each other from our line of sight, either partially or fully



t3- t2 = 0.019 * period = 1264 sec

This is the duration for which the smaller star is entirely in front of the bigger star.
